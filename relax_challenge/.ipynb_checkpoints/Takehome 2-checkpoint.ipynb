```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
users = pd.read_csv(r'takehome_users.csv', encoding='iso-8859-1')
```

```python
users.head()
```

```python
users.rename(columns={'object_id':'user_id'}, inplace=True)
```

```python
users.drop(columns=['name', 'email'], inplace=True)
```

```python
engage = pd.read_csv('takehome_user_engagement.csv')
```

```python
engage.head()
```

```python
engage.info()
```

```python
engage['time_stamp'] = pd.to_datetime(engage['time_stamp'])
```

```python
engage.info()
```

```python
engage.set_index('time_stamp', drop=True, inplace=True)
engage.head()
```

```python
engageq = engage.reset_index()
```

```python
engage['user_id'].nunique()
```

```python
engage.sort_values(axis=0, by='time_stamp', inplace=True)
```

```python
engage['three'] = engage.groupby('user_id')['visited'].transform(lambda s: s.rolling('7D').count())
```

```python
adopted_users = engage[engage['three']>=3]['user_id'].unique()
```

```python
users['user_id'].nunique()
```

```python
users['adopted'] = users['user_id'].apply(lambda x: x in adopted_users)
```

```python
users.head(21)
```

```python
adopted_users.sort()
adopted_users
```

```python
users.info()
```

```python
users.fillna(0, inplace=True)
```

```python
users.info()
```

```python
users
```

```python
users['creation_time']=pd.to_datetime(users['creation_time'])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ohe = OneHotEncoder(drop='first')
colt = ColumnTransformer([('source', ohe, ['creation_source'])])
```

```python
ones = colt.fit_transform(users).toarray()
```

```python
ones = pd.DataFrame(ones, columns=colt.get_feature_names())
```

```python
ones
```

```python
users = users.drop(columns=['creation_source']).merge(ones, left_index=True, right_index=True)
```

```python
users.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestClassifier()
params = {'n_estimators':[10, 50, 100, 200], 'criterion':['gini', 'entropy'], 'max_depth':[None, 2, 5, 10]}
gs = GridSearchCV(estimator=rf, param_grid=params)
```

```python
Xcols = users.drop(columns=['user_id', 'adopted', 'creation_time', 'last_session_creation_time']).columns
X = users.drop(columns=['user_id', 'adopted', 'creation_time', 'last_session_creation_time']).to_numpy()
y = users['adopted'].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
gs.fit(X_train, y_train)
```

```python
gs.cv_results_
```

```python
gs.best_params_, gs.best_score_
```

```python
gs.score(X_test, y_test)
```

```python
rf2 = RandomForestClassifier(criterion='gini', max_depth=2, n_estimators=10)
```

```python
rf2.fit(X_train, y_train)
rf2.score(X_test, y_test)
```

```python
Xcols=list(Xcols)
Xcols
```

```python
features = rf2.feature_importances_
features = pd.DataFrame(features.reshape(1,8), columns=Xcols)
```

```python
features=features.T
features.columns = ['importance']
features.sort_values(by='importance', ascending=False, inplace=True)
```

```python
features
```

```python
sns.barplot(features.T)
plt.show()
```

```python
y_pred = rf2.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
```

```python
cf = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cf, annot=True, cmap="YlGnBu", fmt='d')
plt.show()
```
